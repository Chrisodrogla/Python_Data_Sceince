```python
from selenium import webdriver
import pandas as pd
import time

# set up TOR and the TOR browser
tor_proxy = "socks5://127.0.0.1:9150"
profile = webdriver.FirefoxProfile()
profile.set_preference('network.proxy.type', 1)
profile.set_preference('network.proxy.socks', '127.0.0.1')
profile.set_preference('network.proxy.socks_port', 9150)
profile.set_preference('network.proxy.socks_remote_dns', True)
profile.update_preferences()
driver = webdriver.Firefox(firefox_profile=profile, executable_path='/path/to/geckodriver')

# navigate to the website
site = 'http://k7kg3jqxang3wh7hnmaiokchk7qoebupfgoik6rha6mjpzwupwtj25yd.onion'
driver.get(site)

# get the total number of pages
total_pages = len(driver.find_elements("xpath", '//tr[@style="height:10px;font-size:17px;font-weight: normal;"]/th/span'))

# extract the company links
Company_Link = []
for page in range(1, total_pages+1):
    site_url = site + '/index.php?page=' + str(page)
    driver.get(site_url)
    onclick_attrs = [elem.get_attribute('onclick') for elem in driver.find_elements("xpath", "//th[contains(@class, 'News')]")]
    for onclick_attr in onclick_attrs:
        Company_Link.append(onclick_attr)
    Company_Link = [link.replace("viewtopic('", "").replace("')", "") for link in Company_Link]
    
    
# visit each URL and extract data
alldetails = []
for url in Company_Link:
    driver.get(site + '/topic.php?id=' + url)
    
    gg = driver.find_elements("xpath", '//div[@style="line-height: 1.15;font-size:16px;"]')[0].text.split("\n")
    
    Company_name = gg[0] # Company Name
    Company_Description = gg[9] # Company Description
    Company_Website = gg[3] # Company link
    Size_of_Data = gg[5].strip('amount of data: ') # Data Size
    Listoflink = (site + '/topic.php?id=' + url)
    
    if len(gg) >= 14:
        Download_Leak_Data = gg[13] # Download Data
    elif driver.find_elements("xpath", '/html/body/table/tbody/tr[2]/th/div/h'):
        Download_Leak_Data = driver.find_elements("xpath", '/html/body/table/tbody/tr[2]/th/div/h')[0].text
    else:
        Download_Leak_Data = 'n/a'
    
    DLS_Website = 'PLAY NEWS'
    DLS_Onion_Link = 'http://k7kg3jqxang3wh7hnmaiokchk7qoebupfgoik6rha6mjpzwupwtj25yd.onion'
    #Company = driver.find_elements("xpath", '/html/body/div[3]/div/div[1]/div/h1')[0].text
    Pass = gg[11] # RaR_Password
    
    tempj = {'Company Names': Company_name,
             'Company Description': Company_Description,
             'Company Website': Company_Website,
             'Size of Data': Size_of_Data,
             'Download Data': Download_Leak_Data,
             'RaR Password': Pass,
             'Full_Info': driver.current_url,
             'DLS Website': DLS_Website,
             'DLS Onion Link': DLS_Onion_Link}
    
    alldetails.append(tempj)
    
df = pd.DataFrame(alldetails)
print(df)
```
