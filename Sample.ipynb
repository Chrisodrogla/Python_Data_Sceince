```python
Product_Name = []
Price = []
Star = []
Rate = []


for i in range(1,21):
 
    website = f"https://www.amazon.com/s?k=headphones&crid=28HGBYKT61A0L&qid=1676689882&sprefix=%2Caps%2C244&ref=sr_pg_{i}"
    
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36 Edg/110.0.1587.46", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    page = requests.get(website, headers=headers)

    soup = BeautifulSoup(page.content, "html.parser")
    
    results = soup.find_all('div',{'class':'a-section a-spacing-small a-spacing-top-small'})
    
    for result in results:

        try:
            Product_Name.append(result.find('span',{'class':'a-size-medium a-color-base a-text-normal'}).text)


        except:
            Product_Name.append('n/a')

        try:
            Price.append(result.find('span',{'class':'a-offscreen'}).text.strip('$'))

        except: 
            Price.append('n/a') 
            
        try:
            Star.append(result.find('span',{'class':'a-size-base'}).text)


        except:
            Star.append('n/a')

        try:
            Rate.append(result.find('span',{'class':'a-size-base s-underline-text'}).text.strip('()'))

        except: 
            Rate.append('n/a')
```

```python
Headsetsearchamazon = pd.DataFrame({'Product Name':Product_Name,'Price':Price,'Rate Out of 5 star':Star,'Customer Rate Count':Rate})
```
