```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = 'https://www.toykingdom.com.ph/search?options%5Bprefix%5D=last&page=1&q=car&type=product'
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
results = soup.find_all('div',{'class':'product-item__info'})
```

```python
results[0].find('a',{'style':'font-weight: bold; height: 45px; display: -webkit-box; -webkit-line-clamp: 2; -webkit-box-orient: vertical; overflow: hidden;'}).get_text()

## Product Name

```

```python
Product_Name = []
Price = []

for result in results:
  
    Product_Name.append(result.find('a',{'style':'font-weight: bold; height: 45px; display: -webkit-box; -webkit-line-clamp: 2; -webkit-box-orient: vertical; overflow: hidden;'}).get_text())
    
    Price.append(result.find('span',{'class':'price'}).text.strip('\nSale price\n'))
```

```python
toy_kingdom_car = pd.DataFrame({'Product Name':Product_Name,'Price':Price})
```

# ALL PAGES SCRAPING

```python
Product_Name = []
Price = []

for i in range(1,55):
    x = '&q=car&type=product'
    y = str(i)+x
    website = 'https://www.toykingdom.com.ph/search?options%5Bprefix%5D=last&page=1&q=car&type=product'+ y
    
    response = requests.get(website)
    
    soup = BeautifulSoup(response.content,'html.parser')
    
    results = soup.find_all('div',{'class':'product-item__info'})
    
for result in results:
    
    try:
        Product_Name.append(result.find('a',{'style':'font-weight: bold; height: 45px; display: -webkit-box; -webkit-line-clamp: 2; -webkit-box-orient: vertical; overflow: hidden;'}).get_text())
  
    
    except:
        Product_Name.append('n/a')
        
    try:
        Price.append(result.find('span',{'class':'price'}).text.strip('\nSale price\n'))
        
    except: 
        Price.append('n/a')
    
```

```python
allpages_view = pd.DataFrame({'Product Name':Product_Name,'Price':Price})
```

```python
Product_Name = []
Price = []


for i in range(1,55):
 
    website = f"https://www.toykingdom.com.ph/search?options%5Bprefix%5D=last&page={i}&q=car&type=product"
    
    response = requests.get(website)
    
    soup = BeautifulSoup(response.content,'html.parser')
    
    results = soup.find_all('div',{'class':'product-item__info'})
    
    for result in results:

        try:
            Product_Name.append(result.find('a',{'style':'font-weight: bold; height: 45px; display: -webkit-box; -webkit-line-clamp: 2; -webkit-box-orient: vertical; overflow: hidden;'}).get_text())


        except:
            Product_Name.append('n/a')

        try:
            Price.append(result.find('span',{'class':'price'}).text.strip('\nSale price\n'))

        except: 
            Price.append('n/a') 
```
