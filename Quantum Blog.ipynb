```python
from selenium import webdriver
import pandas as pd
import time
from tqdm import tqdm

# set up TOR and the TOR browser
tor_proxy = "socks5://127.0.0.1:9150"
profile = webdriver.FirefoxProfile()
profile.set_preference('network.proxy.type', 1)
profile.set_preference('network.proxy.socks', '127.0.0.1')
profile.set_preference('network.proxy.socks_port', 9150)
profile.set_preference('network.proxy.socks_remote_dns', True)
profile.update_preferences()
driver = webdriver.Firefox(firefox_profile=profile, executable_path='/path/to/geckodriver')

# navigate to the website
site = 'http://quantum445bh3gzuyilxdzs5xdepf3b7lkcupswvkryf3n7hgzpxebid.onion/'
driver.get(site)

Listoflink = []
Dls_Links = driver.find_elements("xpath", "//div[@class='blog-post-content']/a[2]")
for Dls_Link in Dls_Links:
    Listoflink.append(Dls_Link.get_attribute('href') or Dls_Link)
    
alldetails = []
for i in tqdm(range(len(Listoflink))):
    driver.get(Listoflink[i])
    DLS_Website = 'Quantum Blog'
    DLS_Onion_Link = 'http://quantum445bh3gzuyilxdzs5xdepf3b7lkcupswvkryf3n7hgzpxebid.onion/'
    
    Company_Name = driver.find_elements("xpath", '/html/body/div[2]/div[2]/div/section[1]/div/div/div[2]/div[1]/div[1]/dl/dd[1]')
    Company_Name = Company_Name[0].text if Company_Name else 'n/a'
    
    Company_Desc = driver.find_elements("xpath", '/html/body/div[2]/div[2]/div/section[1]/div/div/div[2]/div[2]/div/p[1]')
    Company_Desc = Company_Desc[0].text if Company_Desc else 'n/a'
    
    Company_Website = driver.find_elements("xpath", '/html/body/div[2]/div[2]/div/section[1]/div/div/div[2]/div[1]/div[1]/dl/dd[2]/a')
    Company_Website = Company_Website[0].get_attribute('href') if Company_Website else 'n/a'

    Download_Data = driver.find_elements("xpath", '/html/body/div[2]/div[2]/div/section[2]/div/div/div/div/div/table/tbody/tr/td[3]/div/div')
    Download_Data = Download_Data[0].get_attribute('href') if Download_Data else Listoflink  # Use Listoflink[i] instead of 'n/a'

    tempj = {
        'Company Names': Company_Name,
        'Company Description': Company_Desc,
        'Company Website': Company_Website,
        'Download Data': Download_Data,
        'Full_Info': Listoflink[i],
        'DLS Website': DLS_Website,
        'DLS Onion Link': DLS_Onion_Link
    }
    
    alldetails.append(tempj)
    
df = pd.DataFrame(alldetails)
print(df)

driver.quit()
```
